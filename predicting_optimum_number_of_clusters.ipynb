```
from sklearn import datasets
```

```
iris=datasets.load_iris()
```

```
data=iris.data
```

```
target=iris.target
```

```
target_names=iris.target_names
```

```
target_names
```

```
from sklearn.cluster import KMeans
```

```
s=[]
for i in range(1,11):
  km=KMeans(n_clusters=i)
  km=km.fit(data)
  s.append(km.inertia_)
```

```
plt.plot(range(1,11),s)
```

```
#elbow bend is found at the point so there we have 3 clusters in conclution.
```

```
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(data)
```

```
y_kmeans
```

```
#cluster based on petal lengh and width
plt.scatter(data[y_kmeans == 0, 0], data[y_kmeans == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(data[y_kmeans == 1, 0], data[y_kmeans == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(data[y_kmeans == 2, 0], data[y_kmeans == 2, 1], 
            s = 100, c = 'green', label = 'Iris-versicolour')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'black', label = 'Centroids')
```

```
#cluster based on sepal length and width
plt.scatter(data[y_kmeans == 0, 2], data[y_kmeans == 0, 3], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(data[y_kmeans == 1, 2], data[y_kmeans == 1, 3], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(data[y_kmeans == 2, 2], data[y_kmeans == 2, 3], 
            s = 100, c = 'green', label = 'Iris-versicolour')
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:,3], 
            s = 100, c = 'black', label = 'Centroids')
```
